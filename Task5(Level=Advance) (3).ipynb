```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 df=pd.read_csv('matches.csv')
df
```

```python
teams=list(df.team1.unique())
teams
```

```python
years=list(df.season.unique())

```

# Most Wins by a Team overall

```python
year=list(df.season.unique())
for team in teams:
    x=df[(df['toss_winner']==team) & (df['winner']==team)]
    plt.bar(team,x.count())
    plt.xticks(rotation='vertical')
    
```

# Man of the Match Players Team Wise

```python

for team in teams:
    w=df[df['winner']==team]
   
    s=dict(w['player_of_match'].value_counts().sort_values(ascending=False)[:10])
    plt.bar(s.keys(),s.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
    plt.title(f"MOM players in {team}")
    plt.xlabel('Player of the match')
    plt.ylabel('Frequencies')
    plt.xticks(rotation='vertical')


    plt.show()
```

# Winner Teams Year Wise

```python
years=list(df.season.unique())
for year in years:
    u=df[df.season==year]
    
    z=dict(u['winner'].value_counts().sort_values(ascending=False))
    plt.bar(z.keys(),z.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
    plt.title(f"Winner Teams in {year}")
    plt.xticks(rotation='vertical')
    plt.xlabel('Winner Teams')
    plt.ylabel('Frequencies')
    plt.show()
    



# years=list(u.season.unique())
# years
```

# Winner Teams Overall

```python
over_all=dict(df.winner.value_counts())
plt.bar(over_all.keys(),over_all.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
plt.title("Winner Teams Overall ")
plt.xticks(rotation='vertical')
plt.xlabel('Winner Teams')
plt.ylabel('Frequencies')
plt.show()
```

```python
winning_toss_winning_match = df[(df['toss_winner'] == df['winner'])]
winning_toss_losing_match = df[(df['toss_winner'] != df['winner'])]
df1 = pd.DataFrame({"Team won the toss":[ winning_toss_winning_match.shape[0],winning_toss_losing_match.shape[0]  ]}, index = ['Won match', 'Lost match']  )
df1
```

# Teams with Batting first and Winning 

```python
bat_first_won=df.loc[(df['toss_decision']=='bat')&(df['toss_winner'] == df['winner']),['winner']]
bat_win=dict(bat_first_won.winner.value_counts())
plt.bar(bat_win.keys(),bat_win.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
plt.title("Teams with Batting first and Winning ")
plt.xticks(rotation='vertical')
plt.xlabel('Winner Teams')
plt.ylabel('Frequencies')
plt.show()
```

# Teams with Feilding first and Winning

```python


field_first_won=df.loc[(df['toss_decision']=='field')&(df['toss_winner'] == df['winner']),['winner']]
field_win=dict(field_first_won.winner.value_counts())
plt.bar(field_win.keys(),field_win.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
plt.title("Teams with Feilding first and Winning ")
plt.xticks(rotation='vertical')
plt.xlabel('Winner Teams')
plt.ylabel('Frequencies')
plt.show()
```

# Maximum matches won by Team by 10 wickets

```python
won_by_10_wics=df.loc[(df['win_by_wickets']==10),'winner']
print(f"Maximum matches won by Team by 10 wickets is {won_by_10_wics.value_counts().idxmax()}")
ten_wics=dict(won_by_10_wics.value_counts())
plt.bar(ten_wics.keys(),ten_wics.values())
plt.title('Maximum matches won by Team by 10 wickets')
plt.xticks(rotation='vertical')
plt.xlabel('Winner Teams')
plt.ylabel('Frequencies')
plt.show()
```

```python
df1=pd.read_csv('deliveries.csv')
df1
```

```python
df2 = df1[df1['dismissal_kind'].notna()]
df2
```

# Dismissals And their Frequencies in IPL


```python
dismissals=dict(df2.dismissal_kind.value_counts())
plt.bar(dismissals.keys(),dismissals.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
plt.title('Dismissals Kind')
plt.xticks(rotation='vertical')
plt.xlabel('Dismissals Kind')
plt.ylabel('Frequencies')

plt.show()
```

# Team-wise Dismissals

```python
batting_teams=list(df2.batting_team.unique())
for team in batting_teams:
    bat_teams=df2[df2['batting_team']==team]
    player_dis=dict(bat_teams.player_dismissed.value_counts().sort_values(ascending=False)[:10])
    plt.bar(player_dis.keys(),player_dis.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
    plt.title(f"{team} Dismissals")
    plt.xticks(rotation='vertical')
    plt.xlabel('Players')
    plt.ylabel('Frequencies')
    plt.show()
    
   
    
```

# Highest  Wicket Takers

```python
bowling_teams=list(df2.bowling_team.unique())
for team in bowling_teams:
    bowl_teams=df2[df2['bowling_team']==team]
    bowlers=dict(bowl_teams.bowler.value_counts().sort_values(ascending=False)[:10])
    plt.bar(bowlers.keys(),bowlers.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
    plt.title(f"{team} Highest Wicket takers")
    plt.xticks(rotation='vertical')
    plt.xlabel('Players')
    plt.ylabel('Frequencies')
    plt.show()
```

# Best Fielding Efforts

```python
fielders=dict(df2.fielder.value_counts().sort_values(ascending=False)[:10])
plt.bar(fielders.keys(),fielders.values(),color=['red', 'blue', 'purple', 'green', 'yellow','magenta'])
plt.title(" Best Fileding Efforts Dismissals")
plt.xticks(rotation='vertical')
plt.xlabel('Players')
plt.ylabel('Frequencies')
plt.show()
```
